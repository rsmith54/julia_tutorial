```julia
function factorial2(n)
    start = 1
    for i = 2:n
        start = start*i
    end
    return(start)
end
    
for j = 1:10
    println(factorial2(j))
end
```

```julia
function binomial_rv(n, p)
    number_of_successes = 0
    for ii = 1:n
        if rand() < p
           number_of_successes += 1 
        end
    end
    return(number_of_successes)
end

dist_sample = []

for trial in 1:100
    push!( dist_sample, binomial_rv( 10, .5 ))
end
```

```julia
function distance( point,  point_prime = [0., 0.])
    x = point[1]
    y = point[2]

    x_prime = point_prime[1]
    y_prime = point_prime[2]

    distance = sqrt( (x - x_prime)^2 + (y - y_prime)^2) 
    return( distance)
end

function rand_2d()
    return( [rand(), rand()])
end

function is_in_unit_circle(point)
    return( distance( point) < 1)
end

n_iterations = 1000000

count_inside_circle = 0
for ii in 1:n_iterations
    point = rand_2d()
    if is_in_unit_circle( point) 
        count_inside_circle += 1
    end
end

# only looks in one quadrant, we need all 4
pi_approx = 4 * count_inside_circle / n_iterations
```

```julia
#Write a program that prints one realization of the following random device:

#Flip an unbiased coin 10 times
#If 3 consecutive heads occur one or more times within this sequence, pay one dollar
#If not, pay nothing

function flip_coin(p_heads = .5) #unbiased by default
    return( rand() < p_heads )
end

function determine_game_loss(seq)
    if length(seq) < 3 
        return("don't pay")
    end
    triplets = seq[1:3]
    
    for ii in seq[1:(length(seq - 2))]
        if triplets[1] && triplets[2] && triplets[3]
            return("Pay one dollar")
        end
    end
end

function play_the_game()
    results = []
    for iter in 1:10
        push!(results, flip_coin())
    end
    return(results)
end

#play the game a bunch of times
for game in 1:1000
    results = play_the_game()
    println( determine_game_loss(results))
end
```

```julia
function calculate_x_t_plus_one(x_t, alpha)
    return( x_t*alpha + rand())
end

function run_time_series(x_0, alpha, nperiods)
    time_series = []
    x_t = x_0

    for iter in 1:nperiods
       x_t_plus_one = calculate_x_t_plus_one(x_t, alpha)
       push!(time_series, x_t_plus_one)
    end
    return(time_series)
end

nperiods = 200 #T
x_0 = 0

results = Dict("0"  => run_time_series(0  , x_0, 200),
               ".8" => run_time_series(.8 , x_0, 200),
               ".98" => run_time_series(.98, x_0, 200),)


plot(1:nperiods, values(results))
```
